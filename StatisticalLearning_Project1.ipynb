```python
import numpy as np
import pandas as pd
```

```python
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data

laliga = pd.read_csv('Laliga.csv', skiprows=1)
print(laliga.head(2))

#Replace all the '-' values to comprise of 0
laliga.replace(to_replace='-', value=0, inplace=True)
laliga.head(10)
```

```python
#Print all the teams which have started playing between 1930-1980
laliga[laliga['Debut'].between('1930','1980',inclusive=True)][['Team', 'Debut']]
```

# Print the list of teams which came Top 5 in terms of points

```python
#### Print the list of teams which came Top 5 in terms of points

laliga['Points'] = laliga['Points'].astype('int') #To convert Object to int type
print(laliga.sort_values('Points', ascending=False).head(5)[['Points', 'Team']])
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
#Using the same function, find the team which has maximum and minimum goal difference.

# Approach 1 using a single method to compute goal difference and to find the MAX and MIN

def Goal_diff_count(team, goals_for, goals_against):
    #Store the goal difference of each team in a dictionary
    team_data = dict()
    for element in range(0, len(list(team))):
        goals_diff = goals_for[element] - goals_against[element]
        team_data[team[element]] = goals_diff
    max_v = max(zip(team_data.values(), team_data.keys()))
    min_v = min(zip(team_data.values(), team_data.keys()))
    print("Max goal difference : {}\n".format(max_v))
    print("Min goal difference : {}\n".format(min_v))
    return team_data

all_team_data = Goal_diff_count(list(laliga['Team']), list(laliga['GoalsFor'].astype('int')), list(laliga['GoalsAgainst'].astype('int')))
print("All team goal difference details :\n{}".format(all_team_data))
```

```python
#Approach 2: To create a new column and store the goal difference, later find the max and min from that column(by using sort)

def goal_difference(goals_for, goals_against):
    goals_difference = goals_for - goals_against
    return goals_difference

laliga['goal_difference'] = goal_difference(laliga['GoalsFor'].astype('int'), laliga['GoalsAgainst'].astype('int'))

print(laliga[['Team', 'goal_difference']])

value = laliga.sort_values(by='goal_difference', ascending=False)[['Team', 'goal_difference']]
print("Max goal difference : {}".format(value.head(1)))
print("Min goal difference : {}".format(value.tail(1)))
```

##### Create a new column with name “Winning Percent” and append it to the data set (5 points)

```python
#Create a new column to compute winning Percent for each row
laliga['Winning Percent'] = round(((laliga['GamesWon'].astype('int') / laliga['GamesPlayed'].astype('int')) * 100), 4)

# Replace the Numerical errors with 0%
laliga['Winning Percent'].fillna(value=0, inplace=True)

#Find out the top 5 teams having the highest Winning percentage
laliga.sort_values(by='Winning Percent', ascending=False).head(5)
```

### Group teams based on their “Best position” and print the sum of their points for all positions (5 points)
