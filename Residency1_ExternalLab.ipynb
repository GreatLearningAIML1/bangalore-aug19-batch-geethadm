### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
response = data["Age"].describe()
#print(response)
q75, q25 = np.percentile(data["Age"], [75, 25])
#print(q75, q25)
iqr_val = q75 - q25
print("The IQR for the input dataset is : {}".format(iqr_val))
```

### Plot the attrition column (Since it is a categorical variable, chose the plot accordingly)

```python
import matplotlib.pyplot as plt

#l = data["Attrition"]
#print(l)
attrition_val = data["Attrition"]
count_val = attrition_val.value_counts()
count_val

attrition_data = {"Yes": 237, "No": 1233}
labels = attrition_data.keys()
values = attrition_data.values()

index = np.arange(len(labels))
plt.bar(index, values)
plt.xlabel('Count', fontsize=15)
plt.ylabel('Attrition', fontsize=15)
plt.xticks(index, labels, fontsize=10, rotation=30)
plt.title('Attrition chart')
plt.show()
```
